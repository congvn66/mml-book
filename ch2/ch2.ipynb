### a.
Check the $⋆$ operation, to see if the operation map from the initial set to itself:
- We let a $⋆$ b = -1:

  $ab + a + b = -1$ ⇒ $a = -1$ or $b = -1$. Because $a$ or $b$ cant be $-1$, the $⋆$ operation will always map from the initial set to itself. (1)

- Check for closure: we can use (1) to show the closure of it.
- Check for associtivity:

  $(a ⋆ b) ⋆ c \\
  = (ab + a + b) ⋆ c \\
  = abc + ac + bc + ab + a + b + c
  $ (I)

  $ a ⋆ (b ⋆ c) \\
  = a ⋆ (bc + b + c) \\
  = abc + ab + ac + bc + b + c + a
  $ (II)

  From (I) and (II) $⇒ (a ⋆ b) ⋆ c = a ⋆ (b ⋆ c)$. Therefore, confirm the associtivity.

- Check for neutral:

  We have: $a ⋆ 0 = a0 + a + 0 = a$ and  $0 ⋆ a = 0a + 0 + a = a$

  Therefore, $e = 0$ is the neutral element.

- Check for inverse:
   
  We can use the neutral element above, which is $e = 0$, we have: $ab + a + b = 0$, solve the equation, we have:

  $b = \frac{1}{a+1} - 1$, similar for $b ⋆ a$.

- Check for commutativity:
  
  Again, easy to check $a ⋆ b = b ⋆ a$, just by using the given formula.


$⇒$ Group $(R\ \{-1\}, ⋆)$ is a Abelian group.

### a.

- Check if the operation is well-defined (because an element of the set has many representations):

  Assume: $\bar a = \bar a'$, $\bar b = \bar b'$, to see if the operation is well-defined or not, we have to verify: $\overline{a + b} = \overline{a' + b'}$

  In the set, $\bar a = \bar a'$ if and only if $a' - a$ is a multiple of n. For example $\overline 1 = \overline 4 = \overline 7$ for $n = 3$. So, we have:
  $a' - a = k_1 \times n$ and $b' - b = k_2 × n$ $⟹ (a' + b') - (a + b) = (k_1 + k_2) × n$. Therefore, $\overline{a + b} = \overline{a' + b'}$, leads to the operation is well-defined

- Check for closure:
  $\bar a \oplus \bar b  = \overline{a + b} \in ℤ_n$.
- Check for associtivity:
  $(\bar a \oplus \bar b) \oplus \bar c = \overline{a+b} \oplus \bar c = \overline{a + b + c} = \bar a \oplus \overline{b+c} = \bar a \oplus (\bar b \oplus \bar c)$
- Check for neutral:
  $e = \overline{0}$ is the neutral element ($\overline{kn}$ cant be because the neutral element must inside $ℤ_n$).
- Check for inverse:
  $\overline{-a}$ is the inverse element.
- Check for commutativity:
  $\bar a \oplus \bar b = \bar b \oplus \bar a = \overline{a + b}$

Therefore,$(ℤ_n, ⊕)$ is an Abelian group.
