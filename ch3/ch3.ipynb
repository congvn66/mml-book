## 3.7

### a.


Consider: $((id_v - \pi) ∘ (id_v - \pi))(x) = (id_v - \pi)(x - \pi(x)) = x - 2\pi(x) + \pi^2(x)$. If $id_v - \pi$ is a projection, $x - 2\pi(x) + \pi^2(x) = x - \pi(x) ⇔ \pi^2(x) = \pi(x) ⇔ \pi$ is a projection.

### b.

We have $Im(id_v - \pi) = \{x - \pi(x) | x \in V\}$, but $\pi(x - \pi(x)) = \pi(x) - \pi^2(x) = 0$. Therefore, $Im(id_v - \pi) ⊆ ker(\pi)$.

On the other hand, given $k \in ker(\pi)$, we have $k - \pi(k) = k  ⇒ k \in Im(id_v - \pi) ⇒ ker(\pi) ⊆ Im(id_v - \pi)$. Thus, $ker(\pi) = Im(id_v - \pi)$.


We have $ker(id_v - \pi) = \{x \in V: x = \pi(x)\} ⇒ ker(id_v - \pi) ⊆ Im(\pi)$

On the other hand, given $x \in Im(\pi) ⇒ ∃y \in V: \pi(y) = x ⇒ (id_v - \pi)(x) = x - \pi(x) = \pi(y) - \pi^2(y) = 0$ $⇒ x \in ker(id_v - \pi) ⇒ Im(\pi) ⊆ ker(id_v - \pi)$.

Thus, $Im(\pi) = ker(id_v - \pi)$.
