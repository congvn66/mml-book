## 4.11

We assume $\lambda \ne 0$ is an eigenvalue of $A^TA$. Therefore, we have:

$$A^TAx = \lambda x$$
$$⇔AA^TAx = \lambda Ax$$

Thus, with $\lambda \ne 0$ is an eigenvalue of $A^TA$ with the eigenvector $x$, $AA^T$ will have the same eigenvalue but different eigenvector.

## 4.12

First, we consider $||Ax||_2^2 = (Ax)^T(Ax) = x^TA^TAx = x^TV\Sigma^TU^TU\Sigma Vx = x^TV\Sigma^T\Sigma V^Tx$

We set $V^Tx = y$. Because $V^T$ is an orthogonal matrix, the length of $x$ will be preserve: $||y||_2^2 = ||x||^2_2$.

Note that, $\Sigma^T\Sigma$ is a square matrix with size $n \times n$. Thus, $||Ax||_2^2 = y^T\Sigma^T\Sigma y = \Sigma_{i = 1}^N(\sigma_i^2y_i^2) \leq \Sigma_{i = 1}^n(\sigma_1^2y_i^2) = \sigma_1^2||y_i||_2^2 ⇒ \frac{||Ax||_2^2}{||y||_2^2} = \frac{||Ax||_2^2}{||x||_2^2} \leq \sigma_1^2 ⇒ \frac{||Ax||_2}{||x||_2} \leq \sigma_1$

Thus, $max_x \frac{||Ax||_2}{||x||_2} = \sigma_1$. The "=" happens if and only if $\Sigma_{i = 1}^n(\sigma_1^2-\sigma_i^2)y_i^2 = 0 ⇔ y = \begin{pmatrix}y_1&0&...&0\end{pmatrix}^T$
